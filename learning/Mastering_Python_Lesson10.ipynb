Iterators â€” The Foundation of Python Loops

An iterator is an object you can loop over, one element at a time.

```python
numbers = [10, 20, 30]
iterator = iter(numbers)

print(next(iterator))  # 10
print(next(iterator))  # 20
print(next(iterator))  # 30
```

Any object that implements:

__iter__()  # returns an iterator
__next__()  # returns next element

is iterable (like lists, tuples, dicts, etc.).

```python
class CountUp:
    def __init__(self, limit):
        self.limit = limit
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.limit:
            self.count += 1
            return self.count
        else:
            raise StopIteration

for num in CountUp(5):
    print(num)
```

Generators â€” Memory-Efficient Iterators

Generators are functions that yield values one at a time instead of returning them all at once.

```python
squares = (x**2 for x in range(5))
print(next(squares))  # 0
print(list(squares))  # [1, 4, 9, 16]
```

Why Generators Matter

Generators donâ€™t store the whole dataset in memory â€” they stream it.

Example: processing millions of database rows or API results efficiently.

```python
def read_large_file(filename):
    with open(filename) as file:
        for line in file:
            yield line.strip()
```

Decorators â€” Functions That Modify Other Functions

Decorators allow you to â€œwrapâ€ functionality around another function without modifying its code.

```python
def greet():
    print("Hello!")

def decorator(func):
    def wrapper():
        print("Before function runs")
        func()
        print("After function runs")
    return wrapper

greet = decorator(greet)
greet()
```

```python
def decorator(func):
    def wrapper():
        print("Before")
        func()
        print("After")
    return wrapper

@decorator
def greet():
    print("Hello!")

greet()
```

```python
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator

@repeat(3)
def say_hello():
    print("Hello!")

say_hello()
```

```python
def log_call(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__} with {args} {kwargs}")
        result = func(*args, **kwargs)
        print(f"{func.__name__} returned {result}")
        return result
    return wrapper

@log_call
def add(a, b):
    return a + b

add(5, 7)
```

```python
import time

# Generator: simulate streaming heart rate data
def heart_rate_stream():
    for bpm in [72, 74, 90, 110, 85]:
        time.sleep(0.5)
        yield bpm

# Decorator: log each value
def log_data(func):
    def wrapper():
        for bpm in func():
            print(f"Heart Rate: {bpm}")
    return wrapper

@log_data
def stream():
    return heart_rate_stream()

stream()
```
