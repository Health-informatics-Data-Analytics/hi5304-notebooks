```python
from sqlalchemy import create_engine, text
import pandas as pd

# 1) Create the database engine (Postgres reachable via localhost:5432)
engine = create_engine(
    "postgresql+psycopg2://hi5304_user:hi5304_password@localhost:5432/hi5304"
)

# 2) Test the connection
with engine.connect() as conn:
    print("Connection test:", conn.execute(text("SELECT 1")).scalar())

# 3) Load patients.csv into a pandas DataFrame
patients = pd.read_csv("../data/patients.csv")

# 4) Write the DataFrame to Postgres as a table named "patients"
patients.to_sql("patients", engine, if_exists="replace", index=False)

# 5) Run a SQL SELECT statement and return results as a DataFrame
query = """
SELECT patient_id, first_name, last_name, sex, zip_code
FROM patients
ORDER BY patient_id;
"""
result = pd.read_sql(query, engine)

result





```
