```python
query = """
SELECT
  p.patient_id,
  p.first_name,
  p.last_name,
  b.reading_date,
  b.systolic,
  b.diastolic,
  b.heart_rate,
  b.source,
  m.medication,
  m.dose,
  m.frequency
FROM patients p
LEFT JOIN bp_readings b ON p.patient_id = b.patient_id
LEFT JOIN medications m ON p.patient_id = m.patient_id
ORDER BY p.patient_id, b.reading_date;
"""
df = pd.read_sql(query, engine)
df.head(20)
```
