## 6. Simple Plot

```
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
plt.title('Example Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
